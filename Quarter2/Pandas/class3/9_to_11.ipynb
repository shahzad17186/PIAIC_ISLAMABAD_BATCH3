```python
#   left_df   right_df on=column(merge based on this)
#how=which joing type you want to use (left,right, inner, outer)
a = df1.merge(df2, on=["CNIC","Account No"], how='inner')
a.head()
```

```python
#   left_df   right_df on=column(merge based on this)
#how=which joing type you want to use (left,right, inner, outer)
a = df1.merge(df2, on=["CNIC","Account No"], how='left')
print(len(a))
a.head()
```

```python
df1 = pd.DataFrame({"Customers NIC": np.random.randint(1,20, size=(20)),
                   "Cus Account":np.random.randint(1,2, size=(20))})


df2 = pd.DataFrame({"CNIC": np.random.randint(1,20, size=(20)),
                   "Account No":np.random.randint(1,2, size=(20)),
                   "Mobile No":np.random.randint(2001,3000, size=(20)),
                  "DOB":pd.date_range('1990-01-01', periods=20),
                  "CustomerName":[chr(x) for x in np.random.randint(65,91, size=(20))],
                   "Gender":np.random.choice(["M","F"], size=(20))})

#Join with different columns name in both table
df1.merge(df2, left_on=["Customers NIC","Cus Account"], right_on=["CNIC","Account No"], how='outer')
```

```python
df1 = pd.DataFrame({"NIC":[1,3,4],"A":[1,3,23]})
df2 = pd.DataFrame({"NIC":[1,2,3],"B":[1,3,23]})
df3 = pd.DataFrame({"NIC":[1,4,7],"C":[1,3,23]})

df = df1.merge(df2, on="NIC", how='left')
df = df.merge(df3, on="NIC",how="left")
df
```

```python
df1 = pd.DataFrame({"NIC":[1,3,4],"A":[1,3,23]})
df2 = pd.DataFrame({"NIC":[1,2,3],"B":[1,3,23]})
df3 = pd.DataFrame({"NIC":[1,4,7],"C":[1,3,23]})

df = df1.merge(df2, on="NIC", how='left').merge(df3, on="NIC",how="left")
df
```
